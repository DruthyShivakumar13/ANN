Checking for outliers

```python
df.plot(kind="box", figsize =(5,5), rot =90, color = "blue")
```

```python
df_numeric = df.select_dtypes(include=[np.number])
# Calculate the IQR for each column
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
iqr = q3 - q1

# Identify outliers as data points with value < Q1 - 1.5*IQR or value > Q3 + 1.5*IQR
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outliers = df_numeric[(df_numeric < lower_bound) | (df_numeric > upper_bound)]
```

```python
outliers.count()
```

```python
for col in df_numeric.columns:
    col_median = df_numeric[col].median()
    df_numeric.loc[df_numeric[col] < lower_bound[col], col] = col_median
    df_numeric.loc[df_numeric[col] > upper_bound[col], col] = col_median
```

```python
outliers_after_impute = df_numeric[(df_numeric < lower_bound) | (df_numeric > upper_bound)]
```

```python
outliers_after_impute.count()
```

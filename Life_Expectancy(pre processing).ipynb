Checking for outliers

```python
def outlier_imputer(df, cols):
    df_imputed = df.copy()
    for col in cols:
        Q1 = df_imputed[col].quantile(0.25)
        Q3 = df_imputed[col].quantile(0.75)
        IQR = Q3 - Q1

        upper_bound = Q3 + 1.5 * IQR
        lower_bound = Q1 - 1.5 * IQR
        
        outliers = (df_imputed[col] > upper_bound) | (df_imputed[col] < lower_bound)
        
        median = df_imputed[col].median()
        df_imputed.loc[outliers, col] = median
    
    return df_imputed
```

Covert data types

```python
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
```

Encode categorical variables

```python
df = pd.get_dummies(df, columns=['Status'])
```

Feature engineering

```python
df.rename(columns={'Life expectancy ': 'Life_Expectancy', 'infant deaths': 'Infant_Deaths'}, inplace=True)

mean_life_expectancy = df.groupby('Country')['Life_Expectancy'].mean().reset_index()
mean_life_expectancy.rename(columns={'Life_Expectancy': 'Mean Life_Expectancy'}, inplace=True)

# merge the mean_life_expectancy DataFrame with the original DataFrame
df = pd.merge(df, mean_life_expectancy, on='Country', how='left')
```

Drop irrelevant columns
